```python
import os
import re
import pandas as pd
import docx

# Define columns you need
columns = ["coin_name", "metal", "weight", "dimensions", "mint", "year", "description"]
data = []

# Define folder path
folder_path = os.path.abspath("/content/drive/MyDrive/Coins")

for file_name in os.listdir(folder_path):
    if file_name.endswith(".docx"):
        try:
            doc = docx.Document(os.path.join(folder_path, file_name))
        except Exception as e:
            print(f"Error processing {file_name}: {e}")
            continue

        # Extract text from paragraphs
        full_text = [para.text for para in doc.paragraphs]

        # Initialize coin information dictionary
        coin_info = {key: "" for key in columns}

        for line in full_text:
            # Normalize the line to lowercase for case-insensitive matching
            line_lower = line.lower()

            # Match fields with regular expressions (case-insensitive)
            if "metal:" in line_lower:
                coin_info["metal"] = re.search(r'metal:\s*(.+)', line, re.IGNORECASE).group(1).strip()
            elif "weight:" in line_lower:
                coin_info["weight"] = re.search(r'weight:\s*(.+)', line, re.IGNORECASE).group(1).strip()
            elif "dimensions:" in line_lower:
                coin_info["dimensions"] = re.search(r'dimensions:\s*(.+)', line, re.IGNORECASE).group(1).strip()
            elif "mint:" in line_lower:
                coin_info["mint"] = re.search(r'mint:\s*(.+)', line, re.IGNORECASE).group(1).strip()
            elif "date of issue:" in line_lower:
                coin_info["year"] = re.search(r'date of issue:\s*(.+)', line, re.IGNORECASE).group(1).strip()
            else:
                # Append non-empty lines to the description
                if line.strip():
                    coin_info["description"] += line.strip() + " "

        # Add parsed coin info to the data list
        data.append(coin_info)

# Create a DataFrame and export to CSV
df = pd.DataFrame(data, columns=columns)
df.to_csv("/content/drive/MyDrive/Coins/coin_data.csv", index=False, encoding="utf-8-sig")
```

```python
import os
import re
import pandas as pd
import docx

def parse_docx_file(docx_path):
    """
    Parses a single .docx file to extract coin info.
    Returns a list of dictionaries, each representing one coin.
    """
    doc = docx.Document(docx_path)

    # Convert paragraphs to a list of text lines
    lines = [para.text.strip() for para in doc.paragraphs if para.text.strip()]

    # We'll store coin data in a list of dicts:
    # Each dict: {"coin_number": ..., "metal": ..., "weight": ..., ... }
    coins = []
    current_coin = {}

    for line in lines:
        # Detect "Coin No. X.X :" or any variant you have
        # Adjust the regex to match your exact doc format
        coin_no_match = re.match(r"Coin\s*No\.*\s*(\d+\.\d+)\s*[:\-]?", line, re.IGNORECASE)
        if coin_no_match:
            # If we already had a coin in progress, push it to the list
            if current_coin:
                coins.append(current_coin)
            # Start a new coin
            current_coin = {
                "coin_number": coin_no_match.group(1),
                "metal": "",
                "weight": "",
                "dimensions": "",
                "mint": "",
                "year": "",
                "description": ""
            }
            continue

        # Detect "Metal: " line
        if line.lower().startswith("metal:"):
            current_coin["metal"] = line.split(":", 1)[-1].strip()
            continue

        # Detect "Weight: " line
        if line.lower().startswith("weight:"):
            current_coin["weight"] = line.split(":", 1)[-1].strip()
            continue

        # Detect "Dimensions: " line
        if line.lower().startswith("dimensions:"):
            current_coin["dimensions"] = line.split(":", 1)[-1].strip()
            continue

        # Detect "Mint: " line
        if line.lower().startswith("mint:"):
            current_coin["mint"] = line.split(":", 1)[-1].strip()
            continue

        # Detect "Year: " line
        if line.lower().startswith("year:"):
            current_coin["year"] = line.split(":", 1)[-1].strip()
            continue

        # If none of the above matched, we treat this line as part of 'description'
        # or you can create an additional pattern if needed
        if current_coin.get("description") is not None:
            current_coin["description"] += " " + line

    # If the last coin is still in progress, append it
    if current_coin:
        coins.append(current_coin)

    return coins


def parse_all_docx_in_folder(folder_path):
    """
    Loops through all .docx files in the specified folder,
    parses them, and returns a combined list of coin dictionaries.
    """
    all_coins = []
    for file_name in os.listdir(folder_path):
        if file_name.lower().endswith(".docx"):
            docx_path = os.path.join(folder_path, file_name)
            coins_in_file = parse_docx_file(docx_path)
            all_coins.extend(coins_in_file)
    return all_coins


# Example usage:
FOLDER_PATH = "/content/drive/MyDrive/Coins"  # Adjust as needed

all_coins_data = parse_all_docx_in_folder(FOLDER_PATH)

# Convert to DataFrame
df = pd.DataFrame(all_coins_data)

# Inspect first rows
print(df.head(10))

# Save to CSV
output_csv = "/content/drive/MyDrive/Coins/coin_data1.csv"  # Adjust path if needed
df.to_csv(output_csv, index=True)
print(f"CSV saved to {output_csv}")
```

```python
import re
import pandas as pd

# Replace with your CSV's actual filename/path
input_csv = "/content/drive/MyDrive/Coins/coin_data1.csv"
output_csv = "/content/drive/MyDrive/Coins/coin_data1_refined.csv"

df = pd.read_csv(input_csv)

# Go row by row
for i, row in df.iterrows():
    desc = row["description"]

    # Split description into lines (based on newlines)
    lines = desc.split("\n")

    # We'll store lines that do NOT match known patterns here,
    # so they remain part of the final 'description'
    leftover = []

    for line in lines:
        line = line.strip()

        # Example fix: If your docx output has "4, Mint:" instead of "4. Mint:"
        # Convert a leading "<digit>," to "<digit>."
        line = re.sub(r"^(\d+),", r"\1.", line)

        # Regex to capture lines like "1. Metal: Silver."
        # Explanation:
        #  ^\d+\.      -> starts with "digit plus dot" (e.g. "1.")
        # \s*(Metal|weight|Dimensions|Mint|Date of issue) -> the label
        # \s*:\s*      -> the colon plus any whitespace
        # (.*)$        -> capture everything else as the value
        pattern = re.compile(r"^\d+\.\s*(Metal|weight|Dimensions|Mint|Date of issue)\s*:\s*(.*)$", re.IGNORECASE)

        match = pattern.match(line)
        if match:
            key = match.group(1).strip().lower()    # e.g., "metal" or "weight"
            value = match.group(2).strip()          # e.g., "Silver." or "0.48 gram."

            # Assign parsed value to the correct column
            if key == "metal":
                df.at[i, "metal"] = value
            elif key == "weight":
                df.at[i, "weight"] = value
            elif key == "dimensions":
                df.at[i, "dimensions"] = value
            elif key == "mint":
                df.at[i, "mint"] = value
            elif key == "date of issue":
                # Some people prefer a separate "date_of_issue" column,
                # but we’ll just fill "year" for simplicity:
                df.at[i, "year"] = value
        else:
            # This line doesn’t match known patterns, so keep it for description
            leftover.append(line)

    # Reconstruct leftover lines for the final 'description'
    df.at[i, "description"] = " ".join(leftover).strip()

# Save the refined CSV
df.to_csv(output_csv, index=False)
print(f"Refined CSV saved as: {output_csv}")
```

```python
from docx import Document

def remove_images_from_docx(input_path, output_path):
    """
    Removes all images from a single .docx file and saves the result to output_path.
    """
    # Open the .docx file
    doc = Document(input_path)

    # Iterate over all the document's elements
    for rel in list(doc.part.rels.values()):
        # Check if the relationship is an image
        if "image" in rel.target_ref:
            # Remove the image
            doc.part.drop_rel(rel.rId)

    # Save the updated document
    doc.save(output_path)
    print(f"Processed and saved: {output_path}")

def process_docx_files(input_folder, output_folder):
    """
    Processes all .docx files in the input folder, removes images,
    and saves the results in the output folder.
    """
    # Ensure the output folder exists
    os.makedirs(output_folder, exist_ok=True)

    for file_name in os.listdir(input_folder):
        if file_name.endswith(".docx"):
            input_path = os.path.join(input_folder, file_name)
            output_path = os.path.join(output_folder, file_name)

            try:
                remove_images_from_docx(input_path, output_path)
            except Exception as e:
                print(f"Error processing {file_name}: {e}")

# Specify input and output paths
input_folder = "/content/drive/MyDrive/Coins/"  # Replace with your folder containing .docx files
output_folder = "/content/drive/MyDrive/Coins/noimg"  # Replace with your desired output folder

# Process all .docx files
process_docx_files(input_folder, output_folder)
```

```python
import os
import re
import pandas as pd
from docx import Document

def parse_coin_docx(file_path):
    """
    Parses a .docx file containing messy coin data and converts it into structured data.
    """
    data = []
    current_king = None
    current_reign = None
    description = None

    # Open the .docx file
    doc = Document(file_path)
    paragraphs = iter(doc.paragraphs)  # Create an iterator from the paragraphs list

    for para in paragraphs:
        text = para.text.strip()

        # Match King/Issuer section
        if re.match(r"^\d+\.\s+[A-Za-z]+", text):
            if current_king:
                # Save data for the previous king/issuer
                data.append({
                    "King/Issuer": current_king,
                    "Reign": current_reign,
                    "Description": description,
                    "Coin Number": None,
                    "Metal": None,
                    "Weight (grams)": None,
                    "Dimensions (mm)": None,
                    "Mint": None,
                    "Date of Issue": None,
                })
            # Parse new king/issuer section
            current_king = re.sub(r"^\d+\.\s+", "", text.split("Reign")[0].strip())
            current_reign = re.search(r"Reign:\s*(.+)", text).group(1).strip() if "Reign" in text else None
            description = None

        # Match Description (if present before coins)
        elif current_king and not description:
            description = text

        # Match individual Coin details
        elif text.startswith("Coin No."):
            coin_info = {"King/Issuer": current_king, "Reign": current_reign, "Description": description}

            # Extract coin number
            coin_info["Coin Number"] = re.search(r"Coin No\.\s*(.+):", text).group(1).strip() if ":" in text else None

            # Process coin-specific details
            while True:
                try:
                    coin_text = next(paragraphs).text.strip()
                except StopIteration:
                    break

                if coin_text.startswith("Coin No."):
                    # Return to outer loop for the next coin
                    paragraphs = iter([coin_text] + list(paragraphs))  # Re-insert unprocessed paragraph
                    break

                if "Metal:" in coin_text:
                    coin_info["Metal"] = re.search(r"Metal:\s*(.+)", coin_text).group(1).strip()
                elif "Weight:" in coin_text:
                    coin_info["Weight (grams)"] = re.search(r"Weight:\s*(.+)", coin_text).group(1).strip()
                elif "Dimensions:" in coin_text:
                    coin_info["Dimensions (mm)"] = re.search(r"Dimensions:\s*(.+)", coin_text).group(1).strip()
                elif "Mint:" in coin_text:
                    coin_info["Mint"] = re.search(r"Mint:\s*(.+)", coin_text).group(1).strip()
                elif "Date of issue:" in coin_text:
                    coin_info["Date of Issue"] = re.search(r"Date of issue:\s*(.+)", coin_text).group(1).strip()

            # Append coin details
            data.append(coin_info)

    return pd.DataFrame(data)

# Specify the path to the .docx file
file_path = "/content/drive/MyDrive/Coins/medieval bengal coin 1 Khilji - 20 Danuz.docx"

# Parse the file and create a DataFrame
coin_data = parse_coin_docx(file_path)

# Save the structured data to a CSV
output_path = "/content/drive/MyDrive/Coins/medieval bengal coin 1 Khilji - 20 Danuz.csv"
coin_data.to_csv(output_path, index=True, encoding="utf-8-sig")

print(f"Structured CSV saved to {output_path}")
```

```python
# ===============================================================
# 🛠 STEP 1: Import Dependencies
# ===============================================================
import os
import concurrent.futures
import torch
import torchvision.transforms as transforms
import numpy as np
import subprocess
from PIL import Image
from tqdm import tqdm
import sys
sys.path.append("/content/ESRGAN")

try:
    from RRDBNet_arch import RRDBNet
    print("✅ ESRGAN module loaded successfully!")
except ModuleNotFoundError as e:
    print("❌ ESRGAN module still not found:", e)


# ===============================================================
# 🔧 STEP 2: Configure Paths
# ===============================================================
input_folder = "/content/drive/MyDrive/Code/Project/Nurul Islam Shangrahashala/Coinage/Shangrahasala coins/Muslim Conquerors"
output_folder = "/content/drive/MyDrive/Code/Project/Upscaled_PNGs"
os.makedirs(output_folder, exist_ok=True)

# ===============================================================
# 🚀 STEP 3: Load ESRGAN Model (GPU-Optimized)
# ===============================================================
model_path = "ESRGAN/models/RRDB_PSNR_x4.pth"
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Define model architecture
model = RRDBNet(in_nc=3, out_nc=3, nf=64, nb=23)
model.load_state_dict(torch.load(model_path, map_location=device), strict=True)
model.eval()
model = model.to(device)

# ===============================================================
# 📈 STEP 4: AI-Based Upscaling Function (Parallel Processing)
# ===============================================================
def upscale_esrgan(input_path, output_path):
    try:
        # Open image
        img = Image.open(input_path).convert("RGB")

        # Convert image to tensor
        img_lr = transforms.ToTensor()(img).unsqueeze(0).to(device)

        # Perform AI upscaling
        with torch.no_grad():
            img_sr = model(img_lr).cpu().squeeze(0).numpy()

        # Convert back to image
        img_sr = np.clip(img_sr * 255.0, 0, 255).astype(np.uint8)
        img_sr = Image.fromarray(np.transpose(img_sr, (1, 2, 0)))

        # Save upscaled image
        img_sr.save(output_path, format="PNG", optimize=True)

        # Optimize with pngquant (Batch Processing)
        subprocess.run(["pngquant", "--quality=65-85", "--speed=1", "--force", "--ext", ".png", output_path], check=True)

        return True
    except Exception as e:
        print(f"❌ Error processing {input_path}: {e}")
        return False

# ===============================================================
# ⏳ STEP 5: Collect PNG Files for Processing
# ===============================================================
png_files = []
for root, _, files in os.walk(input_folder):
    for filename in files:
        if filename.lower().endswith(".png"):
            input_path = os.path.join(root, filename)
            rel_path = os.path.relpath(input_path, input_folder)
            output_path = os.path.join(output_folder, rel_path)

            os.makedirs(os.path.dirname(output_path), exist_ok=True)
            png_files.append((input_path, output_path))

# ===============================================================
# 🚀 STEP 6: Process PNGs in Parallel with Improved Progress Bar
# ===============================================================
converted_count = 0

with concurrent.futures.ThreadPoolExecutor() as executor:
    with tqdm(total=len(png_files), desc="🔄 Upscaling PNGs", unit="img") as pbar:
        futures = {executor.submit(upscale_esrgan, *args): args for args in png_files}
        for future in concurrent.futures.as_completed(futures):
            result = future.result()
            converted_count += int(result)
            pbar.update(1)

print(f"\n🎉 Done! Upscaled {converted_count}/{len(png_files)} images in '{output_folder}'.")
```
